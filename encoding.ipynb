# FEATURE ENCODING

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load Dataset
df = sns.load_dataset('tips')
df
```

# LABEL ENCODING ON `TIME`

```python
# Check the count of each value in the column
df['time'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder

# Call LabelEncode
LE = LabelEncoder()

# Note-REPRESENTATION: { DINNER = 0 | LUNCH = 1 }  

df['encoded_time'] = LE.fit_transform(df['time'])
df.head()
```

```python
# Check the represenatation in Label Encoded form:
df['encoded_time'].value_counts()
```

# ORDINAL ENCODING ON `DAY`

```python
# Check the count of each value in the column
df['day'].value_counts()
```

```python
# In Ordinal Coding, the order is been assigned on which the value will be shown on the dataset

# Note-Representation : { THUR = 0.0 | FRI = 1.0 | SAT = 2.0 | SUN = 3.0 }

OE = OrdinalEncoder(categories=[['Thur','Fri','Sat','Sun']])
df['encoded_day'] = OE.fit_transform(df[['day']])
df.head()
```

```python
# Check the represenatation in Label Encoded form:
df['encoded_day'].value_counts()    # It is according to the given Order.
```

```python
# Check the count of each value in the column
df['sex'].value_counts()
```

# ONE HOT ENCODING

```python
OHE =OneHotEncoder()

# [1,0] : HERE FEMALE IS PRESENT AND MALE IS ABSENT.
# [0,1] : HERE FEMALE IS ABSENT AND MALE IS PRESENT.

OHE.fit_transform(df[['sex']]).toarray()
```

# EXAMPLE OF `ONE-HOT ENCODING`

```python
# LOAD A DATASET OF TTIANIC
titanic = sns.load_dataset('titanic')

# Spare is a matrix in which most of the elements are zero.
ohe = OneHotEncoder(sparse_output=False)
embarked_ohe = ohe.fit_transform(titanic[['embark_town']])

embarked_ohe_df = pd.DataFrame(embarked_ohe, columns = ohe.get_feature_names_out( ['embark_town']))

# reset_index : ye index ko reset kr deta hn for making both the df's in a single page.
titanic = pd.concat([titanic.reset_index(drop=True),embarked_ohe_df.reset_index(drop=True)], axis=1)

titanic.head()
```
